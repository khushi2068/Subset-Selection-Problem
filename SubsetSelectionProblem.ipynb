```python
import random as r
s={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
subsets=set()
while(len(subsets)<100):
 subset=r.sample(s,5)
#  print(subset)
 if sum(subset)==0:
  subsets.add(tuple(subset))

print(subsets)
```
